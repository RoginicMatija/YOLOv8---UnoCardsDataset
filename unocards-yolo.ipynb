```python
from ultralytics import YOLO
model = YOLO()
```

```python
#epochs = [10,25,100,500]
epochs = [500]
train_results = []
val_results = []
for e in epochs:
    model = YOLO()
    result = model.train(data = "/kaggle/input/unocardsdataset/data.yaml", epochs = e, imgsz = 416, verbose = False)
    train_results.append(result)
    vresult = model.val(data = "/kaggle/input/unocardsdataset/data.yaml", imgsz = 416, plots = True)
    val_results.append(vresult)
```
